### Calculate & Plot Correlations

```python
from ggplot import *
%matplotlib inline

ggplot(df_cars, aes(x='hp', y='weight')) +\
  geom_point() + labs(title="Car-Attributes", x="Horsepower", y="Weight")
```

```python
df_cars.plot(x="hp",y="weight",kind="scatter")
```

### Coding Excrcise
1. Calculate the correlation between Rear Axle Ratio & the Width
2. Plot & verify

```python
print('%2.3f' % df_cars.RARatio.corr(df_cars.width, method="pearson"))
print('%2.3f' % df_cars.RARatio.corr(df_cars.width, method="spearman"))
#-0.453
#-0.244
```

```python
from ggplot import *
%matplotlib inline
ggplot(df_cars, aes(x='RARatio', y='width')) +\
  geom_point() + labs(title="Car-Attributes", x="RA Ratio", y="Width")
```
